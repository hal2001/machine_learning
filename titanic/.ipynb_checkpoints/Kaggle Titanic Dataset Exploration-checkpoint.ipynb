## Data Understanding
To understand the data I think it is best to visualize it. Summary statistics and other rolled-up information obfuscates the structure of the data somewhat, while visualization can show a clearer picture of the data. Literally. Both summary stats and visualizations are critical for a complete understanding.

### Load data and assign column names

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style("whitegrid")
dark_gray = '#2e2e2e'
```

```python
df = pd.read_csv('train.csv')
df.columns = ['p_id',
              'survived',
              'p_class',
              'name',
              'sex',
              'age',
              'n_sib_sp',
              'n_parents',
              'ticket',
              'price',
              'cabin',
              'port']
df.head(3)
```

### Check for missing values

```python
df.count()
```

This shows that **age** and **cabin** are missing values. I prefer to visualize this to quickly see large deficiencies.

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(x=list(df), y=df.count(), color=dark_gray, ax=ax)
ax.set(title='Count of Entries per Category', ylabel='Count')
```

In my opinion, this is much quicker to see that **age** and **cabin** are missing values. Obviously **cabin** is missing a significant amount.

*Hypothesis: Cabin will have little bearing on model quality*

### Get statistical summary of numerical data

```python
df.describe()
```

### Look at distributions of each category

```python
df.boxplot()
```

### Look for correlations

```python
g = sns.PairGrid(df)
g.map_upper(plt.scatter, color=dark_gray)
g.map_lower(sns.kdeplot, cmap="Greys", shade=True)
g.map_diag(sns.kdeplot, color=dark_gray, lw=3, legend=False)
```
